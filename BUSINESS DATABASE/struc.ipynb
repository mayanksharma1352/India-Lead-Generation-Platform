```python
import pandas as pd
```

```python
df1 = pd.read_csv("all_india_business_data.csv")
df2 = pd.read_csv("backup_all_india_business_data.csv")
df3 = pd.read_csv("backup_all_india_business.csv")
df4 = pd.read_csv("business_database_cleaned.csv")

df = pd.concat([df1, df2, df3, df4])
```

```python
df.shape
```

```python
df = df[
    df["Phone"].notna() & 
    (df["Phone"].astype(str).str.strip() != "")
    ]
df.shape
```

```python
df = df[
    df["Name"].notna()
    ]

df = df[
    df["Name"].str.strip() != ""
    ]
df.shape
```

```python
df = df.drop_duplicates(
    subset=["Name", "Phone"],
    keep="first"
)
df.shape
```

```python

df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]","",regex=True)
    .str.strip()
)

display(df[["Phone"]].head())
```

```python

df.to_csv(
    "business_database_cleaned.csv",
    index=False
)
```
