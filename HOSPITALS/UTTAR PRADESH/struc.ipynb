```python
import pandas as pd
```

```python
df = pd.read_csv("uttar_pradesh_hospitals.csv")
```

```python
df.shape
```

```python
df = df[
    df["Phone"].notna() & 
    (df["Phone"].astype(str).str.strip() != "")
    ]
df.shape
```

```python
df = df[
    df["Hospital Name"].notna()
    ]

df = df[
    df["Hospital Name"].str.strip() != ""
    ]
df.shape
```

```python

df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]","",regex=True)
    .str.strip()
)

display(df[["Phone"]].head())
```

```python

df.to_csv(
    "uttar_pradesh_hospitals_cleaned.csv",
    index=False
)
```

```python
df1 = pd.read_csv("uttar_pradesh_hospitals_cleaned.csv")

df2 = pd.read_csv("uttar_pradesh_hospitals.csv")
```

```python
final_df = pd.concat(
    [df1, df2],
    ignore_index=True
)
```
