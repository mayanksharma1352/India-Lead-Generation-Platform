```python
import pandas as pd
```

```python
df = pd.read_csv("bihar_hospitals.csv")
```

```python
df.shape
```

```python
df = df[
    df["Phone"].notna() &
    (df["Phone"].astype(str).str.strip() != "")
]
```

```python
df.shape
```

```python


df = df[
    df["Hospital Name"].notna()
]

df = df[
    df["Hospital Name"].str.strip() != ""
]
```

```python
df.shape
```

```python

df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]", "", regex=True)  # remove symbols/icons
    .str.strip()
)

display(df[["Phone"]].head())
```

```python

df.to_csv(
    "bihar_hospitals_cleaned.csv",
    index=False
)
```
