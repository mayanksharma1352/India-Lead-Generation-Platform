```python
 import pandas as pd
```

```python
df = pd.read_csv("andhra_pradesh_hospitals.csv")
```

```python
df.shape
```

```python
duplicate_count =df.duplicated(
    subset=["Hospital Name", "Google Maps URL"]
).sum()
duplicate_count
```

```python
df = df[
    df["Phone"].notna() &
    (df["Phone"].astype(str).str.strip() != "")
]
```

```python
df.shape
```

```python

df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]", "", regex=True)  # remove symbols/icons
    .str.strip()
)


display(df[["Phone"]].head())
```

```python

df.to_csv(
    "andhra_pradesh_hospital_cleaned.csv",
    index=False
)
```
