```python
import pandas as pd
```

```python
df = pd.read_csv("telangana_schools.csv")
```

```python
df.shape
```

```python
duplicate_count = df.duplicated(
    subset=["School Name", "Google Maps URL"]
).sum()
duplicate_count
```

```python
df = df[
    df["Phone"].notna() & 
    (df["Phone"].astype(str).str.strip() != "")
    ]
df.shape
```

```python
df = df[
    df["School Name"].notna()
    ]

df = df[
    df["School Name"].str.strip() != ""
    ]
df.shape
```

```python

df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]","",regex=True)
    .str.strip()
)

display(df[["Phone"]].head())
```

```python

df.to_csv(
    "telangana_schools_cleaned.csv",
    index=False
)
```
