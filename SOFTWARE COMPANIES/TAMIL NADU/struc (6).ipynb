```python
import pandas as pd
```

```python
df1 = pd.read_csv("tamil_nadu_companies.csv")
df2 = pd.read_csv("tamil_nadu_companies1.csv")
df3 = pd.read_csv("tamil_nadu_companies2.csv")

df = pd.concat([df1,df2,df3], ignore_index=True)
```

```python
df.shape
```

```python
df = df[
    df["Phone"].notna() &
    (df["Phone"].astype(str).str.strip() != "")
    ]
df.shape
```

```python
df = df[
    df["Company Name"].notna()
    ]
df = df[
    df["Company Name"].str.strip() != ""
    ]
df.shape
```

```python
df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]","",regex=True)
    .str.strip()
)
display(df[["Phone"]].head())
```

```python
df.to_csv(
    "tamil_nadu_companies_cleaned.csv",
    index=False
)
```
