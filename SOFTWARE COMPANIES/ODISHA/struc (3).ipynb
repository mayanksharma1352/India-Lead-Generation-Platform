```python
import pandas as pd
```

```python
# Read CSV files
df1 = pd.read_csv("odisha_companies.csv")
df2 = pd.read_csv("odisha_companies1.csv")

# Combine rows
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.shape
```

```python
df = df[
    df["Phone"].notna() &
    (df["Phone"].astype(str).str.strip() != "")
    ]
df.shape
```

```python
df = df[
    df["Company Name"].notna()
    ]
df = df[
    df["Company Name"].str.strip() != ""
    ]
df.shape
```

```python
df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]","",regex=True)
    .str.strip()
)
display(df[["Phone"]].head())
```

```python
df.to_csv(
    "odisha_companies_cleaned.csv",
    index=False
)
```
