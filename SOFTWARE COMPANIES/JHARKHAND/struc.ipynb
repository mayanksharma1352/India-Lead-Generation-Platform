```python
import pandas as pd
```

```python
df = pd.read_csv("jharkhand_companies.csv")
```

```python
df.shape
```

```python
df = df[
    df["Phone"].notna() & 
    (df["Phone"].astype(str).str.strip() != "")
    ]
df.shape
```

```python
df = df[
    df["Company Name"].notna()
    ]

df = df[
    df["Company Name"].str.strip() != ""
    ]
df.shape
```

```python

df["Phone"] = (
    df["Phone"]
    .astype(str)
    .str.replace(r"[^\d\s]","",regex=True)
    .str.strip()
)

display(df[["Phone"]].head())
```

```python

df.to_csv(
    "jharkhand_companies_cleaned.csv",
    index=False
)
```
